```python
# ----------------------------
# Train/Validation/Test Split with image validation
# ----------------------------
import os
import numpy as np
import pandas as pd
from PIL import Image
from tensorflow.keras.preprocessing.image import ImageDataGenerator

IMG_HEIGHT = 224
IMG_WIDTH = 224
BATCH_SIZE = 32

dataset_path = "/content/dataset/tomatodataset"

# Get classes
classes = [d for d in os.listdir(dataset_path) if os.path.isdir(os.path.join(dataset_path, d))]
print("Classes:", classes)

# Function to check if an image is valid
def is_valid_image(path):
    try:
        img = Image.open(path)
        img.verify()
        return True
    except:
        return False

# Split dataset into train/val/test
train_files, val_files, test_files = [], [], []

for cls in classes:
    cls_path = os.path.join(dataset_path, cls)
    images = [os.path.join(cls_path, img) for img in os.listdir(cls_path)
              if not img.startswith('.') and is_valid_image(os.path.join(cls_path, img))]

    np.random.shuffle(images)
    n_total = len(images)
    n_train = int(0.7 * n_total)
    n_val = int(0.2 * n_total)

    train_files += [(img, cls) for img in images[:n_train]]
    val_files   += [(img, cls) for img in images[n_train:n_train+n_val]]
    test_files  += [(img, cls) for img in images[n_train+n_val:]]

print(f"Train: {len(train_files)}, Val: {len(val_files)}, Test: {len(test_files)}")

# ----------------------------
# ImageDataGenerators (no augmentation)
# ----------------------------
train_datagen = ImageDataGenerator(rescale=1./255)
val_test_datagen = ImageDataGenerator(rescale=1./255)

def create_generator(file_list, datagen):
    X = [f[0] for f in file_list]
    y = [f[1] for f in file_list]
    return datagen.flow_from_dataframe(
        dataframe=pd.DataFrame({'filename': X, 'class': y}),
        x_col='filename',
        y_col='class',
        target_size=(IMG_HEIGHT, IMG_WIDTH),
        class_mode='categorical',
        batch_size=BATCH_SIZE,
        shuffle=True
    )

# Create generators
train_generator = create_generator(train_files, train_datagen)
val_generator   = create_generator(val_files, val_test_datagen)
test_generator  = create_generator(test_files, val_test_datagen)
```

```python
# ----------------------------
# Ensemble Model: EfficientNetB0 + InceptionV3
# ----------------------------
import tensorflow as tf
from tensorflow.keras.applications import EfficientNetB0, InceptionV3
from tensorflow.keras.layers import Input, Dense, GlobalAveragePooling2D, Concatenate
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam

NUM_CLASSES = len(classes)  # 8 classes

input_layer = Input(shape=(IMG_HEIGHT, IMG_WIDTH, 3))

# EfficientNetB0 branch
b0 = EfficientNetB0(weights=None, include_top=False)(input_layer)
b0 = GlobalAveragePooling2D()(b0)

# InceptionV3 branch
iv3 = InceptionV3(weights=None, include_top=False)(input_layer)
iv3 = GlobalAveragePooling2D()(iv3)

# Concatenate features from both models
combined = Concatenate()([b0, iv3])
output = Dense(NUM_CLASSES, activation='softmax')(combined)

ensemble_model = Model(inputs=input_layer, outputs=output)

# Compile model
ensemble_model.compile(
    optimizer=Adam(learning_rate=1e-4),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

ensemble_model.summary()
```

```python
# ----------------------------
# Train the ensemble model
# ----------------------------
EPOCHS = 60  # You can reduce this to 10-20 for testing first

history = ensemble_model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=EPOCHS,
    verbose=1
)

# Save training history for later analysis
import pickle
with open("/content/ensemble_history.pkl", "wb") as f:
    pickle.dump(history.history, f)
```

```python
# ----------------------------
# Count images per class for Train, Validation, and Test
# ----------------------------

from collections import Counter

def count_images_per_class(file_list):
    classes = [cls for _, cls in file_list]
    return dict(Counter(classes))

train_counts = count_images_per_class(train_files)
val_counts   = count_images_per_class(val_files)
test_counts  = count_images_per_class(test_files)

print("ğŸ“Š Images per class:")

print("\n-- Train Set --")
for cls, count in train_counts.items():
    print(f"{cls}: {count} images")

print("\n-- Validation Set --")
for cls, count in val_counts.items():
    print(f"{cls}: {count} images")

print("\n-- Test Set --")
for cls, count in test_counts.items():
    print(f"{cls}: {count} images")
```

```python
# ----------------------------
# Evaluate model on test set
# ----------------------------
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Get true labels and predictions
y_true = []
y_pred = []

# Class index mapping
class_indices = test_generator.class_indices
inv_class_indices = {v: k for k, v in class_indices.items()}

# Reset generator before predicting
test_generator.reset()

for i in range(len(test_generator)):
    X_batch, y_batch = test_generator[i]
    y_true_batch = np.argmax(y_batch, axis=1)
    y_pred_batch = np.argmax(ensemble_model.predict(X_batch), axis=1)

    y_true.extend(y_true_batch)
    y_pred.extend(y_pred_batch)

# Classification report
print("ğŸ“Š Classification Report:")
print(classification_report(y_true, y_pred, target_names=list(class_indices.keys())))

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=list(class_indices.keys()), yticklabels=list(class_indices.keys()), cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# Overall accuracy
accuracy = np.sum(np.array(y_true) == np.array(y_pred)) / len(y_true)
print(f"\nOverall Accuracy: {accuracy*100:.2f}%")
```

```python
from PIL import Image
import os
import shutil

def resize_and_copy(src, dst, size=(224,224)):
    os.makedirs(dst, exist_ok=True)

    if os.path.isdir(src):
        # src is a folder
        for f in os.listdir(src):
            file_path = os.path.join(src, f)
            if os.path.isfile(file_path):
                try:
                    img = Image.open(file_path).convert("RGB")
                    img = img.resize(size)
                    img.save(os.path.join(dst, f))
                except:
                    print(f"Skipping {file_path}")
    elif os.path.isfile(src):
        # src is a single image
        f = os.path.basename(src)
        try:
            img = Image.open(src).convert("RGB")
            img = img.resize(size)
            img.save(os.path.join(dst, f))
        except:
            print(f"Skipping {src}")
    else:
        print(f"Path does not exist: {src}")
```
